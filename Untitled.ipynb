```python
import os, random
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns

# Reproducibility: fix every source of randomness
SEED = 42
random.seed(SEED)                 # Python RNG (augmentation choices)
np.random.seed(SEED)              # NumPy RNG
torch.manual_seed(SEED)           # CPU tensor RNG (weight init, shuffling)
torch.cuda.manual_seed_all(SEED)  # GPU tensor RNG

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Running on: {DEVICE}")

# Global plot styling, every report figure inherits this professional look
sns.set_theme(style="whitegrid", context="notebook")
plt.rcParams.update({"figure.dpi": 110, "savefig.dpi": 300,
                     "font.family": "DejaVu Sans", "axes.titleweight": "bold"})
FIG_DIR  = "outputs/figures"     if os.path.isdir("outputs") else "../outputs/figures"
CKPT_DIR = "outputs/checkpoints" if os.path.isdir("outputs") else "../outputs/checkpoints"
```

```python
class FruitDataset(Dataset):
    """
    Custom PyTorch Dataset for the Fruits Fresh/Rotten dataset.
    Folder name -> integer label;  image file -> normalized tensor.
    """
    def __init__(self, data_dir, transform=None):
        self.data_dir = data_dir
        self.transform = transform
        self.image_paths = []   # full path of every image
        self.labels = []        # parallel list: labels[i] matches image_paths[i]

        # sorted() guarantees identical class->index mapping on every
        # machine/run (os.listdir order is OS-dependent!)
        self.classes = sorted(
            d for d in os.listdir(data_dir)
            if os.path.isdir(os.path.join(data_dir, d))
        )
        self.class_to_idx = {cls: i for i, cls in enumerate(self.classes)}

        for cls in self.classes:
            cls_dir = os.path.join(data_dir, cls)
            for fname in os.listdir(cls_dir):
                if fname.lower().endswith((".png", ".jpg", ".jpeg")):
                    self.image_paths.append(os.path.join(cls_dir, fname))
                    self.labels.append(self.class_to_idx[cls])

    def __len__(self):
        # DataLoader uses this to compute batches per epoch
        return len(self.image_paths)

    def __getitem__(self, idx):
        # Called lazily by the DataLoader for each sampled index
        image = Image.open(self.image_paths[idx]).convert("RGB")  # force 3 channels
        label = self.labels[idx]
        if self.transform:
            image = self.transform(image)   # PIL -> tensor [3,224,224]
        return image, label
```

```python
# ImageNet channel statistics — the pretrained backbone expects inputs
# normalized with these exact values
IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD  = [0.229, 0.224, 0.225]

# TRAIN pipeline: augmentation ON (overfitting defense #1)
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),            # backbone's expected input size
    transforms.RandomHorizontalFlip(p=0.5),   # orientation is irrelevant for fruit -> free extra data
    transforms.RandomRotation(20),            # camera-angle invariance
    transforms.ColorJitter(brightness=0.2,    # lighting robustness; kept MILD because
                           contrast=0.2,      # color IS the rot signal — aggressive hue
                           saturation=0.15),  # shifts could flip the true label
    transforms.ToTensor(),                    # PIL [0,255] HWC -> float [0,1] CHW
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

# VAL/TEST pipeline: NO augmentation — evaluation must be deterministic
eval_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])
```

```python
# Auto-detect the data folder whether the notebook runs from
# the project root or from notebooks/
DATA_DIR = "data" if os.path.isdir("data") else "../data"
assert os.path.isdir(f"{DATA_DIR}/train"), f"data/train not found relative to {os.getcwd()}"

BATCH_SIZE = 32

# Two views of the SAME train folder with DIFFERENT transforms:
# the train split must see augmentation, the val split must NOT.
full_train_aug   = FruitDataset(f"{DATA_DIR}/train", transform=train_transform)
full_train_clean = FruitDataset(f"{DATA_DIR}/train", transform=eval_transform)
test_dataset     = FruitDataset(f"{DATA_DIR}/test",  transform=eval_transform)

# 85/15 train/val split done on the INDEX list with a seeded generator,
# so both views split identically and reproducibly
n_total = len(full_train_aug)
n_val   = int(0.15 * n_total)
indices = torch.randperm(n_total, generator=torch.Generator().manual_seed(SEED))

train_dataset = torch.utils.data.Subset(full_train_aug,   indices[n_val:].tolist())
val_dataset   = torch.utils.data.Subset(full_train_clean, indices[:n_val].tolist())

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,
                          num_workers=4, pin_memory=True, persistent_workers=True)
val_loader   = DataLoader(val_dataset,   batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=4, pin_memory=True, persistent_workers=True)
test_loader  = DataLoader(test_dataset,  batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=4, pin_memory=True, persistent_workers=True)

CLASSES = full_train_aug.classes
print("Classes:", CLASSES)
print("Class -> index mapping:", full_train_aug.class_to_idx)
print(f"Train: {len(train_dataset)} | Val: {len(val_dataset)} | Test: {len(test_dataset)}")
```

```python
from torchvision import models

def build_model(num_classes, dropout_p=0.3):
    """ResNet18 pretrained on ImageNet, head surgically replaced for our task."""
    model = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)

    # Original head: Linear(512 -> 1000). Replace with Dropout + Linear(512 -> num_classes).
    # Dropout zeroes 30% of features during training -> regularizes the fresh head.
    in_features = model.fc.in_features          # = 512, read instead of hardcoding
    model.fc = nn.Sequential(
        nn.Dropout(p=dropout_p),
        nn.Linear(in_features, num_classes)
    )
    return model

model = build_model(num_classes=len(CLASSES)).to(DEVICE)

total_params     = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Total params: {total_params:,} | Trainable: {trainable_params:,}")
print(f"New classification head: Linear(512 -> {len(CLASSES)})")
```

```python
EPOCHS        = 15
LEARNING_RATE = 1e-4     # low LR: fine-tune good weights, don't destroy them
WEIGHT_DECAY  = 1e-4     # L2 regularization -> overfitting defense
PATIENCE      = 4        # early-stopping patience

# CrossEntropyLoss = LogSoftmax + NLL in one numerically stable op.
# Takes raw logits [B,9] + integer labels [B]; penalizes -log(prob of true class).
criterion = nn.CrossEntropyLoss()

# Adam: per-parameter adaptive learning rate from running gradient mean+variance.
optimizer = torch.optim.Adam(model.parameters(),
                             lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)

# Cosine annealing: LR glides smoothly from 1e-4 toward ~0 over training.
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS)

print("Hyperparameters set. Ready to train.")
```

```python
from tqdm.auto import tqdm

def train_one_epoch(model, loader, criterion, optimizer):
    model.train()                          # Dropout ON, BatchNorm uses batch stats
    running_loss, correct, total = 0.0, 0, 0
    for images, labels in tqdm(loader, leave=False, desc="train"):
        images, labels = images.to(DEVICE), labels.to(DEVICE)
        optimizer.zero_grad()                  # 1. clear accumulated grads
        outputs = model(images)                # 2. forward -> logits [B,9]
        loss = criterion(outputs, labels)      # 3. scalar loss
        loss.backward()                        # 4. backprop (autodiff)
        optimizer.step()                       # 5. update weights
        running_loss += loss.item() * images.size(0)
        correct += (outputs.argmax(1) == labels).sum().item()
        total += labels.size(0)
    return running_loss / total, correct / total

@torch.no_grad()                           # no grad tracking -> faster eval
def evaluate(model, loader, criterion):
    model.eval()                           # Dropout OFF, BatchNorm uses running stats
    running_loss, correct, total = 0.0, 0, 0
    for images, labels in loader:
        images, labels = images.to(DEVICE), labels.to(DEVICE)
        outputs = model(images)
        loss = criterion(outputs, labels)
        running_loss += loss.item() * images.size(0)
        correct += (outputs.argmax(1) == labels).sum().item()
        total += labels.size(0)
    return running_loss / total, correct / total
```

```python
import copy, time

history = {"train_loss": [], "val_loss": [], "train_acc": [], "val_acc": [], "lr": []}
best_val_loss, best_weights, epochs_no_improve = float("inf"), None, 0

start = time.time()
for epoch in range(1, EPOCHS + 1):
    tr_loss, tr_acc = train_one_epoch(model, train_loader, criterion, optimizer)
    va_loss, va_acc = evaluate(model, val_loader, criterion)
    history["lr"].append(optimizer.param_groups[0]["lr"])
    scheduler.step()

    history["train_loss"].append(tr_loss); history["val_loss"].append(va_loss)
    history["train_acc"].append(tr_acc);   history["val_acc"].append(va_acc)

    marker = ""
    if va_loss < best_val_loss:
        best_val_loss = va_loss
        best_weights = copy.deepcopy(model.state_dict())   # snapshot best epoch
        epochs_no_improve = 0
        marker = "  <- best, saved"
    else:
        epochs_no_improve += 1
        if epochs_no_improve >= PATIENCE:
            print(f"Early stopping at epoch {epoch} (no val improvement for {PATIENCE} epochs)")
            break

    print(f"Epoch {epoch:02d}/{EPOCHS} | train loss {tr_loss:.4f} acc {tr_acc:.4f} "
          f"| val loss {va_loss:.4f} acc {va_acc:.4f}{marker}")

print(f"\nTraining time: {(time.time()-start)/60:.1f} min")

model.load_state_dict(best_weights)        # restore BEST, not last (possibly overfit)
torch.save({"model_state": best_weights, "classes": CLASSES},
           f"{CKPT_DIR}/best_model.pth")
print(f"Best val loss {best_val_loss:.4f} | model saved to {CKPT_DIR}/best_model.pth")
```

```python
from sklearn.metrics import (classification_report, confusion_matrix,
                             accuracy_score, roc_curve, auc)
from sklearn.preprocessing import label_binarize

model.eval()
all_labels, all_preds, all_probs = [], [], []

with torch.no_grad():
    for images, labels in tqdm(test_loader, desc="evaluating test set"):
        images = images.to(DEVICE)
        logits = model(images)                       # raw scores [B, 9]
        probs  = torch.softmax(logits, dim=1)        # -> probabilities [B, 9]
        preds  = probs.argmax(dim=1)                 # predicted class index

        all_labels.append(labels.numpy())           # ground truth (stayed on CPU)
        all_preds.append(preds.cpu().numpy())
        all_probs.append(probs.cpu().numpy())        # need full prob vector for ROC

all_labels = np.concatenate(all_labels)             # shape [N]
all_preds  = np.concatenate(all_preds)              # shape [N]
all_probs  = np.concatenate(all_probs)              # shape [N, 9]

test_acc = accuracy_score(all_labels, all_preds)
print(f"Test set size: {len(all_labels)} | Overall test accuracy: {test_acc:.4f}")
```

```python
# We isolate ONE variable: data augmentation. Fixed epoch budget (no early
# stopping) so both convergence curves are the same length and directly comparable.
ABLATION_EPOCHS = 12

def set_seed(seed=SEED):
    """Reset every RNG so both runs start from an identical state."""
    random.seed(seed); np.random.seed(seed)
    torch.manual_seed(seed); torch.cuda.manual_seed_all(seed)

# NO-augmentation training loader over the SAME train images/indices.
# full_train_clean already applies eval_transform (resize+totensor+normalize, no aug).
train_indices      = indices[n_val:].tolist()
train_noaug        = torch.utils.data.Subset(full_train_clean, train_indices)
train_loader_noaug = DataLoader(train_noaug, batch_size=BATCH_SIZE, shuffle=True,
                                num_workers=4, pin_memory=True, persistent_workers=True)

print(f"Aug train images:    {len(train_dataset)}")
print(f"No-aug train images: {len(train_noaug)}  (same images, augmentation disabled)")

def run_experiment(name, train_loader_x, epochs=ABLATION_EPOCHS):
    """Train a fresh model for a fixed number of epochs; return final model + history."""
    set_seed(SEED)                                          # identical init + shuffling
    net  = build_model(num_classes=len(CLASSES)).to(DEVICE)
    crit = nn.CrossEntropyLoss()
    opt  = torch.optim.Adam(net.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)
    sch  = torch.optim.lr_scheduler.CosineAnnealingLR(opt, T_max=epochs)
    hist = {"train_loss": [], "val_loss": [], "train_acc": [], "val_acc": []}
    print(f"\n=== Training [{name}] for {epochs} epochs ===")
    for ep in range(1, epochs + 1):
        tl, ta = train_one_epoch(net, train_loader_x, crit, opt)
        vl, va = evaluate(net, val_loader, crit)
        sch.step()
        hist["train_loss"].append(tl); hist["val_loss"].append(vl)
        hist["train_acc"].append(ta);  hist["val_acc"].append(va)
        print(f"[{name}] epoch {ep:02d}/{epochs} | train acc {ta:.4f} | val acc {va:.4f}")
    return net, hist
```

```python
# Trains TWO models from scratch. On your RTX 4060
model_aug,   hist_aug   = run_experiment("WITH aug",    train_loader)         # augmented
model_noaug, hist_noaug = run_experiment("WITHOUT aug", train_loader_noaug)   # clean

@torch.no_grad()
def test_accuracy(net):
    net.eval(); correct = total = 0
    for images, labels in test_loader:
        images, labels = images.to(DEVICE), labels.to(DEVICE)
        preds = net(images).argmax(1)
        correct += (preds == labels).sum().item(); total += labels.size(0)
    return correct / total

acc_aug_test   = test_accuracy(model_aug)
acc_noaug_test = test_accuracy(model_noaug)
print(f"\nTEST accuracy  WITH aug:    {acc_aug_test:.4f}")
print(f"TEST accuracy  WITHOUT aug: {acc_noaug_test:.4f}")
```

```python
ep_range = range(1, ABLATION_EPOCHS + 1)
fig, axes = plt.subplots(1, 2, figsize=(16, 5.5))

# Left: validation accuracy convergence (the headline comparison)
axes[0].plot(ep_range, hist_aug["val_acc"],   "o-",  color="#2E8B57", lw=2, label="With Augmentation")
axes[0].plot(ep_range, hist_noaug["val_acc"], "s--", color="#C0392B", lw=2, label="Without Augmentation")
axes[0].set_title("Validation Accuracy Convergence", fontsize=13)
axes[0].set_xlabel("Epoch"); axes[0].set_ylabel("Validation Accuracy")
axes[0].legend(); axes[0].grid(alpha=0.3)

# Right: generalization gap = train acc - val acc (overfitting indicator)
gap_aug   = [t - v for t, v in zip(hist_aug["train_acc"],   hist_aug["val_acc"])]
gap_noaug = [t - v for t, v in zip(hist_noaug["train_acc"], hist_noaug["val_acc"])]
axes[1].plot(ep_range, gap_aug,   "o-",  color="#2E8B57", lw=2, label="With Augmentation")
axes[1].plot(ep_range, gap_noaug, "s--", color="#C0392B", lw=2, label="Without Augmentation")
axes[1].axhline(0, color="gray", lw=0.8)
axes[1].set_title("Generalization Gap (Train Acc − Val Acc)", fontsize=13)
axes[1].set_xlabel("Epoch"); axes[1].set_ylabel("Accuracy Gap")
axes[1].legend(); axes[1].grid(alpha=0.3)

plt.suptitle("Ablation Study — Effect of Data Augmentation", fontsize=15, fontweight="bold")
plt.tight_layout()
plt.savefig(f"{FIG_DIR}/fig8_ablation_augmentation.png", bbox_inches="tight")
plt.show()
```

```python
final_gap_aug   = hist_aug["train_acc"][-1]   - hist_aug["val_acc"][-1]
final_gap_noaug = hist_noaug["train_acc"][-1] - hist_noaug["val_acc"][-1]

ablation_df = pd.DataFrame({
    "Configuration":   ["With Augmentation", "Without Augmentation"],
    "Final Train Acc": [hist_aug["train_acc"][-1],  hist_noaug["train_acc"][-1]],
    "Final Val Acc":   [hist_aug["val_acc"][-1],    hist_noaug["val_acc"][-1]],
    "Test Acc":        [acc_aug_test,               acc_noaug_test],
    "Train-Val Gap":   [final_gap_aug,              final_gap_noaug],
}).round(4)
print(ablation_df.to_string(index=False))

fig, ax = plt.subplots(figsize=(12, 2.2)); ax.axis("off")
tbl = ax.table(cellText=ablation_df.values, colLabels=ablation_df.columns,
               cellLoc="center", loc="center")
tbl.auto_set_font_size(False); tbl.set_fontsize(11); tbl.scale(1, 2)
for (r, c), cell in tbl.get_celld().items():
    if r == 0:
        cell.set_facecolor("#3B6EA5"); cell.set_text_props(color="white", weight="bold")
plt.title("Ablation Summary — Data Augmentation", fontsize=13, fontweight="bold", pad=12)
plt.savefig(f"{FIG_DIR}/fig9_ablation_table.png", bbox_inches="tight", dpi=300)
plt.show()

print(f"\nTest accuracy delta (aug − no-aug): {acc_aug_test - acc_noaug_test:+.4f}")
print(f"Overfitting-gap delta (no-aug − aug): {final_gap_noaug - final_gap_aug:+.4f}  "
      f"(positive => augmentation reduced overfitting)")
```
